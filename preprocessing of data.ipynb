```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

```python
data=pd.read_csv("creditcard.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
fraud=data[data['Class']==1]
valid=data[data['Class']==0]
outlier_fraction=len(fraud)/len(valid)
print("outlier_fraction:{}".format(outlier_fraction))
print("fraud_transactions:{}".format(len(fraud)))
print("valid_transactionss:{}".format(len(valid)))
```

# the outlier fraction is 0.17 which is very low so the data is very unbalanced

```python
print("amount details of fraud transactions")
fraud.Amount.describe()
```

```python
print("amount details of valid transactions")
valid.Amount.describe()
```

# average amount of fraud transactions is more

```python
correlation_matrix=data.corr()
print(correlation_matrix)
```

```python
fig=plt.figure(figsize=(20,10))
sns.heatmap(correlation_matrix,vmax=.9,square=True)
plt.show()
```
